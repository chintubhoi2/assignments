```python
import unittest
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
def count_pos_tags(paragraph):
    tokens = word_tokenize(paragraph)
    tags = pos_tag(tokens)
    counts = {'Noun': 0, 'Verb': 0, 'Adverb': 0}

    for word, tag in tags:
        if tag.startswith('N'):
            counts['Noun'] += 1
        elif tag.startswith('V'):
            counts['Verb'] += 1
        elif tag.startswith('R'):
            counts['Adverb'] += 1

    return counts


paragraph1 = "The quick brown fox jumps over the lazy dog.hello world"
count_pos_tags(paragraph1)
```
